# Genre Analysis
This notebook compares Bechdel test scores across different genres to see which genres tend to have better or worse representation of women.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'data/Bechdel_IMDB_Merge0524.csv'
data = pd.read_csv(file_path)

filtered_data = data.dropna(subset=['bechdelRating'])

melted_data = filtered_data.melt(id_vars=['title', 'year', 'imdbid', 'id', 'bechdelRating'], value_vars=['genre1', 'genre2', 'genre3'], var_name='genre_type', value_name='genre')
melted_data = melted_data.dropna(subset=['genre'])

# Plotting the distribution of Bechdel scores across genres
plt.figure(figsize=(14, 7))
sns.boxplot(x='genre', y='bechdelRating', data=melted_data)
plt.xticks(rotation=90)
plt.title('Bechdel Test Scores Across Genres')
plt.xlabel('Genre')
plt.ylabel('Bechdel Test Score')
plt.show()
```

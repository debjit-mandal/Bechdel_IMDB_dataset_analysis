# Analyzing Women's Representation and Movie Length
This notebook analyzes whether women's representation, as measured by the Bechdel test scores, is proportional to the movie's length.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'data/Bechdel_IMDB_Merge0524.csv'
data = pd.read_csv(file_path)

filtered_data = data.dropna(subset=['runtimeMinutes', 'bechdelRating'])

filtered_data['runtimeMinutes'] = pd.to_numeric(filtered_data['runtimeMinutes'], errors='coerce')
filtered_data = filtered_data.dropna(subset=['runtimeMinutes'])

# Plotting the relationship between movie length and Bechdel score
plt.figure(figsize=(12, 6))
sns.boxplot(x='bechdelRating', y='runtimeMinutes', data=filtered_data)
plt.title('Relationship between Movie Length and Bechdel Test Score')
plt.xlabel('Bechdel Test Score')
plt.ylabel('Movie Length (Minutes)')
plt.show()
```

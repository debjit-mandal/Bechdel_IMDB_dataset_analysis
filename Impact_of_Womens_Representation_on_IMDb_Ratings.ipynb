# Impact of Women's Representation on IMDb Ratings
This notebook investigates whether there is a correlation between a movie's Bechdel test score and its IMDb rating.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'data/Bechdel_IMDB_Merge0524.csv'
data = pd.read_csv(file_path)

filtered_data = data.dropna(subset=['imdbAverageRating', 'bechdelRating'])

# Plotting the relationship between IMDb rating and Bechdel score
plt.figure(figsize=(10, 6))
sns.boxplot(x='bechdelRating', y='imdbAverageRating', data=filtered_data)
plt.title('Relationship between IMDb Rating and Bechdel Test Score')
plt.xlabel('Bechdel Test Score')
plt.ylabel('IMDb Rating')
plt.show()

# Calculating correlation between IMDb rating and Bechdel score
correlation = filtered_data['imdbAverageRating'].corr(filtered_data['bechdelRating'])
print(f'Correlation between IMDb rating and Bechdel score: {correlation}')
```

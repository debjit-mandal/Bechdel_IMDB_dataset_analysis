# Analyzing Sexism in Sci-Fi Movies
This notebook analyzes whether sci-fi movies are more likely to have lower scores on the Bechdel test compared to movies of other genres.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'data/Bechdel_IMDB_Merge0524.csv'
data = pd.read_csv(file_path)

filtered_data = data.dropna(subset=['bechdelRating'])

filtered_data['is_sci_fi'] = filtered_data.apply(lambda row: 'Sci-Fi' in [row['genre1'], row['genre2'], row['genre3']], axis=1)

# Plotting the distribution of Bechdel scores for sci-fi vs non-sci-fi movies
plt.figure(figsize=(12, 6))
sns.countplot(x='bechdelRating', hue='is_sci_fi', data=filtered_data)
plt.title('Bechdel Test Scores: Sci-Fi vs Non-Sci-Fi Movies')
plt.xlabel('Bechdel Test Score')
plt.ylabel('Number of Movies')
plt.legend(title='Is Sci-Fi')
plt.show()
```

# Trends in Women's Representation Over Time
This notebook analyzes how the representation of women in movies (based on Bechdel test scores) has changed over the decades.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'data/Bechdel_IMDB_Merge0524.csv'
data = pd.read_csv(file_path)

filtered_data = data.dropna(subset=['bechdelRating'])

filtered_data['decade'] = (filtered_data['year'] // 10) * 10

decade_avg_score = filtered_data.groupby('decade')['bechdelRating'].mean().reset_index()

# Plotting the average Bechdel score over the decades
plt.figure(figsize=(14, 7))
sns.lineplot(x='decade', y='bechdelRating', data=decade_avg_score, marker='o')
plt.title('Average Bechdel Test Score Over the Decades')
plt.xlabel('Decade')
plt.ylabel('Average Bechdel Test Score')
plt.grid(True)
plt.show()
```
